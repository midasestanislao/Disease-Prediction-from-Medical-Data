## Transforming text to integers

```python
df['Heart_Disease'].value_counts(normalize = True)
df['Heart_Disease'].unique()
```

```python
df['General_Health'].value_counts(normalize = True)
df['General_Health'].unique()
```

```python
columns_to_replace = ['Heart_Disease', 'Exercise', 'Skin_Cancer', 'Other_Cancer', 'Depression', 'Diabetes', 'Arthritis', 'Smoking_History']
df[columns_to_replace] = df[columns_to_replace].replace({'No': 0, 'Yes': 1})
df = df[df['Diabetes'].isin([0, 1])]
df['Diabetes'].unique()
```

```python
sns.catplot(data=df, y='Exercise', x='Sex', kind='bar')
plt.show()
```

```python
df.groupby("Sex")["Diabetes"].mean()
```

```python
df["Sex"] = df["Sex"].replace({"Male": 1, "Female": 2})
df["General_Health"] = df["General_Health"].replace({"Poor": 1, "Fair": 2, "Good": 3, "Very Good": 4, "Excellent": 5})
```

```python
df['Age_Category'].unique()
```

```python
df['Age_Category'].value_counts().plot(kind = 'bar')
```

```python
df['Heart_Disease'].value_counts().plot(kind = 'bar')
```

```python
## Select features and target
X = df.drop(columns = ['Heart_Disease', "Checkup", "Age_Category"], axis = 1)
y = df['Heart_Disease']
print(X)
```

```python
# Implementation of K-Nearest Neighbors Algorithm
from sklearn.neighbors import KNeighborsClassifier # K-Nearest Neighbors
from sklearn.linear_model import LogisticRegression # Logistic Regression
from sklearn.naive_bayes import GaussianNB # Naive Bayes
from sklearn.ensemble import RandomForestClassifier # Random Forest Classifier
from sklearn.tree import DecisionTreeClassifier # Decision Tree Classifier
from sklearn.ensemble import GradientBoostingClassifier

# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split 
```

```python
# Splitting the dataset into the Training set and Test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# Define the model
algorithm = KNeighborsClassifier(n_neighbors =  3, metric = 'minkowski', p = 2)
logistic = LogisticRegression()
nb = GaussianNB()
forest = RandomForestClassifier(n_estimators = 100, criterion = 'entropy', random_state = 42, max_depth = 5)
tree = DecisionTreeClassifier(criterion = 'entropy', random_state = 42, max_depth = 5)
gb = GradientBoostingClassifier(n_estimators = 100, learning_rate = 1.0, max_depth = 1, random_state = 42)
```

```python
# Fit the model
algorithm.fit(X_train, y_train)
logistic.fit(X_train, y_train)
# svm.fit(X_train, y_train)
nb.fit(X_train, y_train)
forest.fit(X_train, y_train)
tree.fit(X_train, y_train)
gb.fit(X_train, y_train)
```

```python
# Predicting the Test set results
y_pred = algorithm.predict(X_test)
y_pred_logistic = logistic.predict(X_test)
y_pred_nb = nb.predict(X_test)
y_pred_forest = forest.predict(X_test)
y_pred_tree = tree.predict(X_test)
y_pred_gb = gb.predict(X_test)
# y_pred
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix : \n", cm)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy of KNN model: ", accuracy)
accuracy = accuracy_score(y_test, y_pred_logistic)
print("Accuracy of linear model: ", accuracy)
accuracy = accuracy_score(y_test, y_pred_nb)
print("Accuracy of naive bayes model: ", accuracy)
accuracy = accuracy_score(y_test, y_pred_forest)
print("Accuracy of random forest model: ", accuracy)
accuracy = accuracy_score(y_test, y_pred_tree)
print("Accuracy of decision tree model: ", accuracy)
accuracy = accuracy_score(y_test, y_pred_gb)
print("Accuracy of decision gradient boost model: ", accuracy)
```
